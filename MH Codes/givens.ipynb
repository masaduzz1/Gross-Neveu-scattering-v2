```python
import numpy as np

def get_coeff(N, J, h, k, x, xp, sigma):
    nk = np.arange(-N+1, N+1, 2)
    coeff = 0.0
    for i in range(len(nk)):
        kp = nk[i] * np.pi / N
        phi_k = np.arctan(J * np.sin(kp) / (h - J * np.cos(kp)))
        sec = 1/np.cos(phi_k)
        gauss = np.exp(-np.abs(kp - k)**2 / (4*sigma**2))
        coeff += np.exp(-1j * kp * (xp - x)) * gauss * sec
    return coeff

# Create Given's rotation matrix
def Giv_Mat(N_s, theta, s):
    mat = np.identity(N_s, dtype=float)
    mat[s-1, s-1] = np.cos(theta)
    mat[s-1, s] = -np.sin(theta)
    mat[s, s-1] = np.sin(theta)
    mat[s, s] = np.cos(theta)
    return mat
```

```python
N = 16
k = -7 * np.pi / N
x = 11
J = 0.5
h = 1
sigma = 0.75

coeffs = np.zeros((N), dtype=complex)
for xp in range(N):
    coeffs[xp] = get_coeff(N, J, h, k, x, xp, sigma)

coeffs = np.array(coeffs)
coeffs /= np.linalg.norm(coeffs)

coeffs
```

```python
betas = [-np.angle(coeff) for coeff in coeffs]
p = np.diag(np.exp(1j * np.array(betas)))

coeffs = p @ coeffs
thetas = []
for i in range(N - 1):
    j = N - 1 - i
    theta = np.arctan(-np.abs(coeffs[j]) / np.abs(coeffs[j-1]))
    thetas.append(theta)
    coeffs = Giv_Mat(N, theta, j) @ coeffs
print(coeffs)
```
